```python
# Remove duplicate rows
df = df.drop_duplicates()
# Confirm duplicates removed
df.duplicated().sum()
```

```python
# Remove rows with missing values
df = df.dropna()
# Confirm missing values removed
df.isnull().sum()
```

```python
# Convert Release_Date to datetime format
df['Release_Date'] = pd.to_datetime(df['Release_Date'])
# Extract year from release date
df['Release_Date'] = df['Release_Date'].dt.year
# Display sample row
df.head(1)
```

```python
# Drop columns not needed for analysis
df = df.drop(columns=['Overview','Poster_Url','Original_Language'])
```

```python
# Convert genre string into list
df['Genre']= df['Genre'].str.split(', ')
# Split genres into multiple rows
df = df.explode('Genre').reset_index(drop=True)
# Show first rows
df.head()
```

```python
# Count frequency of each genre
df['Genre'].value_counts()
```

```python
# Dataset shape after preprocessing
df.shape
```

```python
import plotly.express as px

# Count how many times each genre appears in the dataset
genre_counts = df['Genre'].value_counts()

# Create a bar chart showing the frequency of each genre
fig = px.bar(
    x=genre_counts.index,
    y=genre_counts.values,
    title="Most Common Movie Genres"
)

# Display the chart
fig.show()
```

```python
# Sort movies by popularity in descending order and select top 10
top_movies = df.drop_duplicates(subset="Title") \
.sort_values(by="Popularity", ascending=False) \
.head(10)
# Create a horizontal bar chart for better readability of movie titles
fig = px.bar(
top_movies,
x="Popularity", # Popularity score on X-axis
y="Title", # Movie titles on Y-axis
orientation="h", # Horizontal bars
title="Top 10 Most Popular Movies"
)
# Display the chart
fig.show()
```

```python
# Remove duplicate movies first (since each movie appears multiple times for different genres)
unique_movies = df.drop_duplicates(subset='Title')
# Count how many unique movies were released each year
movies_per_year = unique_movies['Release_Date'].value_counts().sort_index()
# Create a line chart to show movie production trend over time
fig = px.line(
x=movies_per_year.index, # Years
y=movies_per_year.values, # Number of movies released
markers=True, # Show points on the line
title="Movies Released Over the Years"
)
# Display the chart
fig.show()
```

```python
# Calculate the average popularity score for each genre
genre_popularity = df.groupby('Genre')['Popularity'].mean().reset_index()
# Create a bar chart showing average popularity of each genre
fig = px.bar(
genre_popularity,
x="Genre", # Genre names
y="Popularity", # Average popularity score
title="Average Popularity by Genre"
)
# Display the chart
fig.show()
```

```python
# Create a histogram to understand how popularity scores are distributed
fig = px.histogram(
df,
x="Popularity", # Popularity values
nbins=30, # Number of bins in the histogram
title="Distribution of Movie Popularity"
)
# Display the chart
fig.show()
```
